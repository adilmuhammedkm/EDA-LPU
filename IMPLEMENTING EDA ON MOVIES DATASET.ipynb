# Dataset Info

id: id of the movies

title: Name of the movies

tagline: Shortnote about the movie

release_date: Release date of the movie 

genres: Defines genre of the movie

belongs_to_collection: Shows if it is a collection of movies or single one

original_language: Language of the movie

budget_musd: Budget of the movie

revenue_musd: Revenue generated using the movie

production_companies: The company that produced the movie

production_countries: Which country the movie was made

vote_count: Who liked the movie

vote_average: Average of the vote count

popularity: How much people know about this movie

runtime: Time duration of the movie

cast: Real Names of characters acted in the movie

cast_size: How many people acted

crew_size: How many people worked for this movie

director: Who directed the movie
