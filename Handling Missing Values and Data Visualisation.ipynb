```python
import pandas as pd
import numpy as np
```

```python
#Example Dataset
data = [55,67,45,68,78,45,85,90,75]
```

```python
#descriptive statistics
mean = np.mean(data)
median = np.median(data)
mode = pd.Series(data).mode()[0]
variance = np.var(data)
std_dev = np.std(data)
```

```python
print(f"Mean:{mean}")
print(f"Median:{median}")
print(f"Mode:{mode}")
print(f"Variance:{variance}")
print(f"standard deviation:{std_dev}")
```

# Missing Values

### Definition:missing data refers to the absence of data points in a dataset. Handling missing data involves identifying these and managing them,such as through imputation or removal

### Mean Imputation :replacing missing values with the mean of the column.
### Median Imputation : replacing missing values with the median.
### Mode Imputation : replacing missing values with the mode.
### Handling Missing data is important as it can impact the resultts of your analysis if not addressed properly.


```python
data = pd.Series([10,20,np.nan,40,50,np.nan,70])
```

```python
#identifying missing data
missing_data_count = data.isna().sum()
print(f"Number of missing values: {missing_data_count}")
```

```python
#mean imputation (filling missing data with mean values)
data_mean_imputed = data.fillna(data.mean())
print("Data after mean imputation:")
print(data_mean_imputed)
```

```python
#median imputation (filling missing data with median values)
data_median_imputed = data.fillna(data.median())
print("Data after median imputation:")
print(data_median_imputed)
```

```python
#mode imputation (filling missing data with mode values)
data_mode_imputed = data.fillna(data.mode())
print("Data after mode imputation:")
print(data_mode_imputed)
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [500,600,700,800,750,680,720,780,850,900,870,920]
```

```python
#Bar charts
plt.figure(figsize=(10,5))
plt.bar(months,sales)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('sales')
plt.show()
```

```python
#Line chart
plt.figure(figsize=(10,5))
plt.plot(months,sales,marker = 'o')
plt.title('Monthly sales trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(7,7))
plt.pie(sales,labels=months, autopct = '%1.1f%%')
plt.title('Sales Distribution')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10,5))
plt.scatter(months, sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales scatter plot')
plt.show
```
